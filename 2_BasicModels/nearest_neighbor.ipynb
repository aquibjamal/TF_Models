```
# Nearest neighbour to classify MNIST images

import numpy as np
import tensorflow as tf

# Import MNIST data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/data", one_hot=True)
```

```
# use of limited MNIST images
Xtr, Ytr = mnist.train.next_batch(5000) # 5000 for training (nn candidates)

Xte, Yte = mnist.test.next_batch(200) #200 for test

# tf Graph input
xtr = tf.placeholder("float", [None, 784])
xte = tf.placeholder("float", [784])

# Nearest neighbour calculation using L1 distances
# calculate L1 distances
distance = tf.reduce_sum(tf.abs(tf.add(xtr, tf.negative(xte))), reduction_indices=1)

# predition : get min distance index (Nearest neighbour)
pred = tf.argmin(distance,0)

accuracy = 0.0

# initailize the global variables
init = tf.global_variables_initializer()
```

```
# Start training
with tf.Session() as sess:
  
  sess.run(init)
  
  # loop over test data
  for  i in range(len(Xte)):
    # Get nearest neighbour
    nn_index = sess.run(pred, feed_dict={xtr:Xtr, xte:Xte[i,:]})
    
    # get nearest neigbour class label and compare to its true label
    print("Test",i, "Prediction",np.argmax(Ytr[nn_index]),
         "True class",np.argmax(Yte[i]))
    
    # calculate accuracy
    if np.argmax(Ytr[nn_index]) == np.argmax(Yte[i]):
      accuracy += 1./len(Xte)
      
  print("Done !")
  print("Accuracy", accuracy)
```
